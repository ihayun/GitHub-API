```python
import requests
import json
from pprint import pprint

from google.colab import userdata
token = userdata.get('ghtoken')


url = 'https://api.github.com/search/repositories?q=sort=stars&order=desc&per_page=20'
headers = {'Authorization': 'Bearer ' + token}

r = requests.get(url, headers=headers)

r_json = json.loads(r.text)
items = r_json['items']
pprint(len(items))
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pd.DataFrame(items).plot.bar()
plt.show()
```
